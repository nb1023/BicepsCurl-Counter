```python
import cv2
import mediapipe as mp
import numpy as np

# gives all the drawing utilities
mp_drawing = mp.solutions.drawing_utils
# importing our pose estimation model
mp_pose = mp.solutions.pose
```

```python
# using webcam -> VIDEO FEED
# grabing or setting up our video capture device... the number is for webcam or device number
cap = cv2.VideoCapture(0)
while cap.isOpened():
#     ret is the return variable
#     frame will the image from our webcam
    ret, frame = cap.read()
#     gives a popup that allows to visualize the particular image
    cv2.imshow('Mediapipe Feed', frame)
#     the above four lines of code -> access the webcam , 
# starting the loop, reading the webcam feed and then visualizing 
# what our webcam is seeing
    
#     break the loop pressing the "q"
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
# to release the wbecam
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
# setup mediapipe instance
# accessing pose estimation model
# what our detection confidence should be also specifying tracking confidence
# for accurate detection increase the value of detection_confidence but for too high it doesn't accurately detect
with mp_pose.Pose(min_detection_confidence =0.5, min_tracking_confidence = 0.5) as pose: 
    while cap.isOpened():
        ret, frame = cap.read()
        
#         Detect stuff and render
# recolor image because when we pass image to mediapipe we don't want it to be in RGB 
# when we get feed using opencv by default it is in BGR
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
#     to save a bunch of memory once we pass it to our pose estimation model
        image.flags.writeable = False
        
#         Make detection
# storing the detections in avariable named results
        results = pose.process(image)
        
#         recoloring back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
#         Render detections
# to draw in particular image
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
#                                   color that we want at particular dots
                                  mp_drawing.DrawingSpec(color = (245, 117, 66), thickness =2, circle_radius = 2),
#                                   connection color
                                  mp_drawing.DrawingSpec(color = (245, 66, 230), thickness =2, circle_radius = 2)

                                 )
        
        cv2.imshow('Mediapipe Feed', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```

### Calculate Angles (shoulder, elbow, wrist)


```python
# calculate angles
def calculate_angle(a,b,c):
    a = np.array(a)  # first point 11
    b = np.array(b)  # mid  point 13
    c = np.array(c)  # end point 15
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle>180.0:
        angle = 360 - angle
        
    return angle
```

```python
shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
```

```python
# to setup the video capturing device i.e webcam in the machine
# 0 a number that represents the webcam 
cap = cv2.VideoCapture(0)

# stage will tell the state of up or down
counter = 0
stage = None

# setup mediapipe instance
# Pose - to access th pose estimation model 
with mp_pose.Pose(min_detection_confidence = 0.5, min_tracking_confidence = 0.5) as pose:

    while cap.isOpened():
    #     frame will give the image from the webcam
        ret, frame = cap.read()
        
#         detect stuff and render
#         Recolor the image in opencv the feed will be taken 
#         in the form of BGR by default but in mediapipe we need to pass
#         in the format of RGB therefore we need to convert it 
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
#     to save the memory once we pass this to our pose estimation model
        image.flags.writeable = False
        
#         making detection
        results = pose.process(image)
        
#         recoloring back to bgr the image
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
#         extract landmarks
        try:
            landmarks = results.pose_landmarks.landmark
            
#             get coordinates
            shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
            elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
            
#             calculate angle
            angle = calculate_angle(shoulder, elbow, wrist)
    
#             visualize angle
# dimensions of webcam - 640, 480
# puText method from cv2 to see the angle and the angle on the screen
# array multiplication method is used
# position of the image/elbow point in the window 
            cv2.putText(image, str(angle), 
                       tuple(np.multiply(elbow, [640, 480]).astype(int)),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA
                       )
    
#             curl counter logic
            if angle>160: 
                stage = "down"
            if angle < 30 and stage == "down":
                stage = "up"
                counter += 1
                print(counter)
                       
        
        except:
            pass
        
#         render curl counter in the frame
#         setup status box
        cv2.rectangle(image, (0,0), (225, 73), (245, 117, 16), -1)
    
#           repetition data
        cv2.putText(image, 'REPS', (15,12),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, str(counter),
                   (10,60),
                   cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2, cv2.LINE_AA)
        
#           STAGE data
        cv2.putText(image, 'STAGE', (65,12),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
        cv2.putText(image, stage,
                   (60,60),
                   cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,255), 2, cv2.LINE_AA)
        
#         render detections
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                 mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                                 )
    
    
    #     it allows to visualize a particular image
        cv2.imshow('Mediapipe Feed', image)
    #     0xFF check which key is pressed in the keyboard
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    # to release the webcam
    cap.release()
    cv2.destroyAllWindows()
```
